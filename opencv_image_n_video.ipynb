## opencv를 이용한 video 처리

```python
!mkdir ./data
!wget -O /content/data/Night_Day_Chase.mp4 https://github.com/chulminkw/DLCV/blob/master/data/video/Night_Day_Chase.mp4?raw=true
```

```python
import cv2

video_input_path = '/content/data/Night_Day_Chase.mp4'
video_output_path = '/content/data/Night_Dat_Chase_Out.mp4'

cap = cv2.VideoCapture(video_input_path)
codec = cv2.VideoWriter_fourcc(*'XVID')

vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
vid_fps = cap.get(cv2.CAP_PROP_FPS)

vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size)

frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print('총 FRAME 갯수 :', frame_cnt, 'FPS :', round(vid_fps), 'FRAME 크기 :', vid_size)
```

```python
import time

green_color = (0, 255, 0)
red_color = (255, 0, 0)

start = time.time()
index = 0

while True :
  hasFrame, img_frame = cap.read()
  if not hasFrame:
    print('더 이상 처리할 Frame이 없습니다.')
    break
  
  index += 1
  print('frame : ', index, '완료')
  cv2.rectangle(img_frame, (300,100,800,400), color = green_color, thickness = 2)
  caption = "Frame : {}".format(index)
  cv2.putText(img_frame, caption, (300,95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 2)

  vid_writer.write(img_frame)

print('write 완료시간 : ', round(time.time() - start, 4))
vid_writer.release()
cap.release()
```

## opencv를 이용한 image 처리

```python
!mkdir ./data2
!wget -O /content/data2/beatles01.jpg https://github.com/chulminkw/DLCV/blob/master/data/image/beatles01.jpg?raw=true
```

```python
#PIL 패키지를 이용하여 이미지 로드

import matplotlib.pyplot as plt
import os
from PIL import Image

PIL_image = Image.open('/content/data2/beatles01.jpg')
print('image tpye : ', type(PIL_image))

plt.figure(figsize=(8,8))
plt.imshow(PIL_image)
```

```python
#skimage(사이킷이미지)로 이미지 로드

from skimage import io

sk_image = io.imread('/content/data2/beatles01.jpg')
print('skimage type : ', type(sk_image), 'skimage shape : ', sk_image.shape)

plt.figure(figsize=(8,8))
plt.imshow(sk_image)
```

```python
#OpenCV로 이미지 로드 - imwrite 방법 사용
#OpenCV는 imread를 하게되면 BGR로 인식을 하게된다. imwrite를 사용하면 BGR > RGB로 다시 바뀌게됨
import cv2

cv2_image = cv2.imread('/content/data2/beatles01.jpg')
cv2.imwrite('/content/data2/beatles01_cv.jpg', cv2_image)
print('cv2_image type : ', type(cv2_image))

img = plt.imread('/content/data2/beatles01_cv.jpg')

plt.figure(figsize=(8,8))
plt.imshow(img)
```

```python
#OpenCV로 이미지 로드 - cvtColor 방법 사용

cv2_image = cv2.imread('/content/data2/beatles01.jpg')
cvt_cv2_image = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)
print('cv_image type : ', type(cvt_cv2_image), 'cv_image shape : ', cvt_cv2_image.shape)

plt.figure(figsize=(8,8))
plt.imshow(cvt_cv2_image)
```
